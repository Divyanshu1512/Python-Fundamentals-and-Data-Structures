3. What are logical operators in Python?
- Logical operators in Python—and, or, and not—are the "decision-makers" used to combine multiple conditional statements. In a practical sense, you use and when you need every single condition to be true (like checking if a price is low and you have enough cash), or when only one of several conditions needs to be met to trigger an action, and not to reverse a result or check if something is false. In your quantitative projects, these are the tools that allow your code to handle complex market rules, ensuring a trade only executes when all your specific criteria align perfectly.

4. Write a Python program to convert user input from string to integer, float,

 7. What is the difference between for and while loops?
 - The primary difference between a for loop and a while loop lies in how they decide to stop. A for loop is generally used when you have a specific collection of items to process or a set number of repetitions in mind, such as iterating through a list of stock tickers or a range of 10 days of data. In contrast, a while loop is used when you don't know exactly how many times the code needs to run and instead want it to continue as long as a certain condition remains true. In your arbitrage project, you would use a for loop to check a fixed list of five currency pairs for price gaps, but you would use a while loop to keep your trading bot running "while" the market is open or "until" your profit target is reached.

8. Write a Python program to calculate the factorial of a number provided by
the user using a while loop.


```python
num = int(input("Enter a number: "))
factorial = 1
while num > 0:
    factorial *= num
    num -= 1
print(f"The factorial is: {factorial}")
```

 9. Why might you use a set instead of a list in Python?
 - A set is used instead of a list in Python primarily for two reasons: uniqueness and speed. Unlike a list, a set automatically removes any duplicate values, ensuring that every element is unique. Furthermore, sets are optimized for "membership testing," meaning they can check if an item exists within them almost instantly, whereas a list must scan every single item one by one.

10. Write a program that checks if a number is positive, negative, or zero.

```python
num = float(input("Enter a number: "))

if num > 0:
    print("The number is Positive")
elif num < 0:
    print("The number is Negative")
else:
    print("The number is Zero")
```
